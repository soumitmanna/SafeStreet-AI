```python
# ==========================================
# Load Experiment Tracker
# ==========================================

import pandas as pd
from pathlib import Path

tracker_path = Path("../reports/experiment_tracker.csv")

experiment_tracker = pd.read_csv(tracker_path)

print("Experiment Tracker Loaded Successfully!")
print()

experiment_tracker
```

```python
# ==========================================
# Add Experiment E04 - XGBoost
# ==========================================

new_experiment = pd.DataFrame([
    {
        "Experiment ID": "E04",
        "Model": "XGBoost",
        "Accuracy": 0.6227,
        "Precision (Macro)": 0.59,
        "Recall (Macro)": 0.43,
        "F1 Score (Macro)": 0.43,
        "Training Time (min)": 0.32,
        "Status": "Candidate",
        "Remarks": "Best accuracy so far with extremely fast training."
    }
])

experiment_tracker = pd.concat(
    [experiment_tracker, new_experiment],
    ignore_index=True
)

experiment_tracker
```

```python
# ==========================================
# Save Updated Experiment Tracker
# ==========================================

experiment_tracker.to_csv(
    tracker_path,
    index=False
)

print("Experiment Tracker Updated Successfully!")
print(f"Saved at: {tracker_path.resolve()}")
```

```python
# ==========================================
# Add Experiment E05 - CatBoost
# ==========================================

new_experiment = pd.DataFrame([
    {
        "Experiment ID": "E05",
        "Model": "CatBoost",
        "Accuracy": 0.6201,
        "Precision (Macro)": 0.59,
        "Recall (Macro)": 0.42,
        "F1 Score (Macro)": 0.42,
        "Training Time (min)": 0.66,
        "Status": "Candidate",
        "Remarks": "Very close to XGBoost but slightly lower accuracy and slower training."
    }
])

experiment_tracker = pd.concat(
    [experiment_tracker, new_experiment],
    ignore_index=True
)

experiment_tracker
```

```python
# ==========================================
# Select Current Champion Model
# ==========================================

# Reset all candidate models
experiment_tracker.loc[
    experiment_tracker["Status"] == "Champion",
    "Status"
] = "Candidate"

# Set XGBoost as the current champion
experiment_tracker.loc[
    experiment_tracker["Model"] == "XGBoost",
    "Status"
] = "Champion"

experiment_tracker[
    ["Experiment ID", "Model", "Accuracy", "Status"]
]
```

```python
# ==========================================
# Save Final Experiment Tracker
# ==========================================

experiment_tracker.to_csv(
    tracker_path,
    index=False
)

print("Experiment Tracker Saved Successfully!")
print(f"Saved to: {tracker_path.resolve()}")
```
